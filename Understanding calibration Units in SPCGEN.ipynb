```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import constants
```

```python
freq = np.linspace(2e9,4e9,201)
T_hot = 3730
T_cold = 2730
cc = constants.c
hh = constants.h
kk = constants.k
T_to_P = lambda T,nu : 2*hh*nu/(np.exp((hh*nu)/(kk*T))-1)
Planckian = lambda T,nu : ((2*hh*nu*nu*nu)/(cc*cc))/(np.exp((hh*nu)/(kk*T))-1)
T_CMB = 2.728 #Kelvi
```

```python
P_hot = T_to_P(T_hot,freq)
P_cold = T_to_P(T_cold,freq)
```

```python
plt.plot(freq/1e9,P_hot,label='P_hot')
plt.plot(freq/1e9,P_cold,label='P_cold')
plt.grid()
plt.ylabel('Power')
plt.xlabel('Freq [GHz]')
plt.legend(loc='best')
plt.show()
```

```python
P_dif = P_hot-P_cold
plt.plot(freq/1e9,P_dif,label='P_dif')
plt.grid()
plt.ylabel('Power')
plt.ylabel('Freq [GHz]')
plt.legend(loc='best')
plt.show()
```

### Uncomment cell below for consistency check of Planckian formula for CMB intensity

```python
test_freq = np.logspace(1,12,50)
CMB_full_intensity = Planckian(T_CMB,test_freq)
plt.semilogx(test_freq,CMB_intensity)
plt.grid()
plt.xlabel('Freq [Hz]')
plt.ylabel('Intensity')
plt.show()
```

```python
CMB_intensity = Planckian(T_CMB,freq)
plt.plot(freq,CMB_intensity)
plt.grid()
plt.ylabel('CMB intensity')
plt.xlabel('Freq [GHz]')
plt.show()

plt.plot(freq,(T_hot-T_cold)/P_dif)
plt.grid()
plt.ylabel('Calibrating factor K/W')
plt.xlabel('Freq [GHz]')
plt.show()


CMB_intensity_RJ = CMB_intensity*(cc*cc/(freq*freq))
T_CMB_derived = CMB_intensity_RJ*(T_hot-T_cold)/P_dif
plt.plot(freq,T_CMB_derived)
plt.grid()
plt.ylabel('Calibrated CMB Temp [K]')
plt.xlabel('Freq [GHz]')
plt.show()
```
